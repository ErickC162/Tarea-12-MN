```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, h, n):
    t_values = [t0]
    y_values = [y0]
    for _ in range(n):
        y0 = y0 + h * f(t0, y0)
        t0 = t0 + h
        t_values.append(t0)
        y_values.append(y0)
    return t_values, y_values

def f_a(t, y):
    return t * np.exp(3 * t) - 2 * y

t0, y0 = 0, 0
h = 0.5
n = int((1 - t0) / h)

t_values_a, y_values_a = euler_method(f_a, t0, y0, h, n)

plt.plot(t_values_a, y_values_a, 'bo-', label='Aproximación')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Método de Euler - Ejercicio 1a')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f3_d(t, y):
    return -5*y + 5*t**2 + 2*t

t0, y0 = 0, 1/3
h = 0.1
n = int((1 - t0) / h)

def euler_method(f, t0, y0, h, n):
    t_values = np.zeros(n + 1)
    y_values = np.zeros(n + 1)
    t_values[0], y_values[0] = t0, y0
    
    for i in range(1, n + 1):
        t_values[i] = t_values[i-1] + h
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
    
    return t_values, y_values

t_values_3d, y_values_3d = euler_method(f3_d, t0, y0, h, n)

plt.plot(t_values_3d, y_values_3d, 'bo-', label='Aproximación')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Método de Euler - Ejercicio 1d')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def taylor_order_2(f, df, t0, y0, h, n):
    t_values = [t0]
    y_values = [y0]

    for i in range(n):
        t = t_values[-1]
        y = y_values[-1]
        
        y_next = y + h * f(t, y) + (h**2 / 2) * df(t, y)
        t_next = t + h

        t_values.append(t_next)
        y_values.append(y_next)

    return t_values, y_values
```

a.

```python
def f_a(t, y):
    return t * np.exp(3 * t) - 2 * y

def df_a(t, y):
    return np.exp(3 * t) * (3 * t + 1) - 2 * f_a(t, y)

t0_a, y0_a, h_a, n_a = 0, 0, 0.5, 2
t_values_a, y_values_a = taylor_order_2(f_a, df_a, t0_a, y0_a, h_a, n_a)
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.plot(t_values_a, y_values_a, 'bo-', label="6(a) h = 0.5")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("6(a) Método de Taylor de orden 2")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def taylor_order_4(f, df1, df2, df3, t0, y0, h, n):
    t_values = [t0]
    y_values = [y0]

    for i in range(n):
        t = t_values[-1]
        y = y_values[-1]

        y_next = (y + h * f(t, y) + 
                  (h**2 / 2) * df1(t, y) + 
                  (h**3 / 6) * df2(t, y) +
                  (h**4 / 24) * df3(t, y))
        t_next = t + h

        t_values.append(t_next)
        y_values.append(y_next)

    return t_values, y_values
```

a.

```python
def f_a(t, y):
    return t * np.exp(3 * t) - 2 * y

def df1_a(t, y):
    return np.exp(3 * t) * (3 * t + 1) - 2 * f_a(t, y)

def df2_a(t, y):
    return 9 * np.exp(3 * t) * (t + 1) - 2 * df1_a(t, y)

def df3_a(t, y):
    return 27 * np.exp(3 * t) - 2 * df2_a(t, y)

t0_a, y0_a, h_a, n_a = 0, 0, 0.5, 2
t_values_a, y_values_a = taylor_order_4(f_a, df1_a, df2_a, df3_a, t0_a, y0_a, h_a, n_a)


plt.figure(figsize=(6, 4))
plt.plot(t_values_a, y_values_a, 'bo-', label="7(a) h = 0.5")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("7(a) Método de Taylor de orden 4")
plt.grid()
plt.legend()
plt.show()
```

```python
def f_b(t, y):
    return 1 + (t - y)**2

def df1_b(t, y):
    return 2 * (t - y) * (1 - f_b(t, y))

def df2_b(t, y):
    return 2 * ((1 - f_b(t, y))**2 - (t - y) * df1_b(t, y))

def df3_b(t, y):
    return -2 * (df1_b(t, y) + (t - y) * df2_b(t, y))

t0_b, y0_b, h_b, n_b = 2, 1, 0.5, 2
t_values_b, y_values_b = taylor_order_4(f_b, df1_b, df2_b, df3_b, t0_b, y0_b, h_b, n_b)

plt.figure(figsize=(6, 4))
plt.plot(t_values_b, y_values_b, 'go-', label="7(b) h = 0.5")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("7(b) Método de Taylor de orden 4")
plt.grid()
plt.legend()
plt.show()
```
